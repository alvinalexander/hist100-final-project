## Analysis of Distictive Words

For our second check in, I wanted to analyze distinctive words, as I thought it would function as an interesting foil in comparing them to our previous results found from our topic modeling exercise.

```python
import pandas
import matplotlib.pyplot as plt
import nltk
import numpy as np
df = pandas.read_csv("../data/bbtop15-lyrics-exploring-tokenized.csv")

#drop rows where the text is missing.
df = df.dropna(subset=['lyrics'])

#view the dataframe
df
```

```python
df['tokens_clean_lower_stopfree'] = df['tokens_clean_lower_stopfree'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

```python
# first imported the function CountVectorizer, then looked at first 100 corresponding words 

from sklearn.feature_extraction.text import CountVectorizer

countvec = CountVectorizer()

sklearn_dtm = countvec.fit_transform(df.tokens_clean_lower_stopfree)
print(sklearn_dtm)
```

```python
print(countvec.get_feature_names()[:100])
```

```python
# created a document term matrix 

dtm_df = pandas.DataFrame(countvec.fit_transform(df.tokens_clean_lower_stopfree).toarray(), columns=countvec.get_feature_names(), index = df.index)
dtm_df
```

```python
#sorted the top referenced words 

print(dtm_df.sum().sort_values(ascending=False))
```

```python
# sort the average of the top referenced words

print(dtm_df.mean().sort_values(ascending=False))
```

### Observation A: On Top Referenced Words

I find it interesting that the top referenced words include "im", "got", and "get". This can be interpreted in accordance with our previous postulations from our topic modeling exercise, as two of our interpreted themes were related to a self-assertion of status (one through expressing power and dominance, the other as an assertion of musical skill). The words "im", "got", and "get" all relay this similar sentiment of asserting status, particularly "get" and "got" expresses ownership dominant status through that means. This hypothesis cannot be confirmed unless compared with lists of the most referenced words from other musical genres. 

I also find it interesting that "like" is high on the list, and I would be curious to know if this is because rappers are making lots of analogies or if this is influenced by the "California accent" whereby in colloquial settings individuals use the word "like" in like every other word. 

```python
#import the function, create a new dtm with cells weighted by the tf-idf score 
# next step would be to look at the highest ranking of weighted words

from sklearn.feature_extraction.text import TfidfVectorizer

tfidfvec = TfidfVectorizer()
dtm_tfidf_df = pandas.DataFrame(tfidfvec.fit_transform(df.tokens_clean_lower_stopfree).toarray(), columns=tfidfvec.get_feature_names(), index = df.index)
dtm_tfidf_df
```

```python
print(dtm_tfidf_df.max().sort_values(ascending=False)[:20])
```

## Observation B: Tf-idf Scores 

I find it interesting that these words manifested as highly weighted content words. That the word "progress" scored so highly is extremely compelling, and I would stipulate that this is because many rap lyrics are represented as narratives of success or struggle. This is something that would be interesting to interpret in the context of the topic modeling exercise, as a possible interpretation of all the different categories could be to view each as a narrative of "progress". For instance, some of the previously-noted themes from our last exercise could also be reframed as: "the progression of gangs, ghettos and violence"; "the progress in appealing for status"; "the progress of musical self-recognition"; "the progression of seduction"; and "the progression of self-expression". I would like to investigate this more in seeing the context of how the word "progress" is being used. 

"La" is another interesting variable, though I find its significance to be less compelling. I am mostly curious as to if this due to the duality of its usage, as the word could be interpreted as "la" as in the city and "la" as in the onomatopoeia sing-song.

Lastly, in considering the list as a sovereign category, I can't help but notice that in assessing the top scoring words, there is another pattern of phrases that emphasize status. Words such as "fame", "amazing", "versace" and (debatably) "la" all emulate testament of an assertion of status in different contexts. This relates to Observation A and Observation B whereby themes of assertion of status from the previous exercise of content modeling hold potential relevance. 

```python
#creat dataset with document index and gender and then merge this into the dtm_tfidf_df

df_gender = df['artist_gender'].to_frame()
merged_df = df_gender.join(dtm_tfidf_df)

merged_df
```

```python
(df['artist_gender'].value_counts())
```

```python
#pull out the highesting scoring tf-idf scores to compare females and males by print the words with the highest tf-idf scores for each genre

dtm_female = merged_df[merged_df['artist_gender']=="Female"]

print("Female")
print(dtm_female.max(numeric_only=True).sort_values(ascending=False)[0:40])
print()

```

```python
dtm_male = merged_df[merged_df['artist_gender']=="Male"]

print("Male")
print(dtm_male.max(numeric_only=True).sort_values(ascending=False)[0:40])
print()
```
