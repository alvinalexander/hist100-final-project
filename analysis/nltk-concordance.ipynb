```python
from ast import literal_eval
import pandas
import matplotlib.pyplot as plt
import numpy as np
import nltk 
from time import time
df = pandas.read_csv("../data/bbtop15-lyrics-tokenized.csv",converters={'tokens_lower':literal_eval, 'tokens_lower_clean':literal_eval } )

#drop rows where the text is missing.
df = df.dropna(subset=['lyrics'])

#view the dataframe
df
```

# NLTK Concordance

Here is how you can use the concordance function. 

Steps:

1. Turn lists of tokens into text-objects using the nltk.Text() function 
2. Call the concordance function on these text objects. eg text_object.concordance('money')

Also note that you must use the tokens in the tokens_lower_clean column. 


Ideas:

subset the data frame by year and apply the function on lyrics from that specific year.

#### You should also take a look at this link to see how you can use the 'concordance' and the 'similar ' function in nltk. http://www.nltk.org/book/ch01.html

### Example code:

```python
#Creat a text object of the first song in our list.

text_object = nltk.Text(df['tokens_lower_clean'][0])
```

```python
# Call the concordance function with the string 'me' as input. 
text_object.concordance('me')
```

```python
text_objects = [nltk.Text(tokens) for tokens in df['tokens_lower_clean'].tolist()]
```

```python
#### Print out all the lyrics that uses the word money in some context

[text.concordance('money') for text in text_objects if text.concordance('money') != None or text.concordance('money') != "No matches"]
```

```python
 [text.similar('money') for text in text_objects if text.similar('money') != None ]
```
