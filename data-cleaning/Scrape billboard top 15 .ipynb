```python
from lxml import html
import requests
import csv
import datetime
from datetime import timedelta

```

```python
##Get all dates of billboard 15
dates = []
start_date = datetime.datetime.strptime("1989-11-04", "%Y-%m-%d")
curr_date = start_date

while curr_date.year < 2017:
    dates.append(curr_date.strftime("%Y-%m-%d"))
    curr_date = curr_date + datetime.timedelta(days=7)
            
print(dates)
#curr_date.month < 13 and curr_date.day < 31
```

```python
def get_page_tree(date):
    page = requests.get('http://www.billboard.com/charts/rap-song/%s'%date)
    tree = html.fromstring(page.content)
    return tree
```

```python
def get_artists(tree):
    artists = tree.xpath('//*[@class="chart-row__artist"]/text()')
    artists = [artist.strip() for artist in artists]
    return artists

def get_songs(tree):
    songs = tree.xpath('//*[@class="chart-row__song"]/text()')
    songs = [song.strip() for song in songs]
    return songs

def get_positions(tree):
    positions = tree.xpath('//*[@class="chart-row__current-week"]/text()')
    positions = [int(position.strip()) for position in positions]
    return positions
    

def get_top15(date):
    top15 = []
    page_tree = get_page_tree(date)
    artists = get_artists(page_tree)
    songs = get_songs(page_tree)
    positions = get_positions(page_tree)
    for i in range(0, len(artists)):
        top15.append([artists[i], songs[i], positions[i], date])
        
    return top15
        
    
    
```

```python
#Get all top 15 from november 1889 to 2016

f = open('../data/final-project-data/bbtop15-1989-2016.csv', 'wt')
try:
    writer = csv.writer(f)
    writer.writerow( ('artist', 'track', 'position', 'date') )
    for date in dates:
        top15 = get_top15(date)
        for song in top15:
            writer.writerow(song)
finally:
    f.close()

```
